```python
import pandas as pd 
import numpy as np
```

```python
train = pd.read_csv('proj_2_train_churn.csv')
test = pd.read_csv('proj_2_test_churn.csv')
submit_ids = test['id']
train
```

```python
train.describe()
```

```python
train = train.drop(['id','category','Fld_408'], axis = 1)
```

```python
col_to_drop = []

for col in train.columns:
    if train[col].min() == train[col].max():
        col_to_drop.append(col)

col_to_drop
```

```python
train = train.drop(col_to_drop, axis = 1)
```

```python
train.isnull().sum().sort_values(ascending=False)
```

```python
train.fillna(-999, inplace=True)
```

```python
train.isnull().sum().sort_values(ascending=False)
```

```python
col_to_drop_test = []

for col in test.columns:
    if test[col].min() == test[col].max():
        col_to_drop_test.append(col)

col_to_drop_test
```

```python
test.drop(col_to_drop_test, axis = 1, inplace = True)
```

```python
test
```

```python
test = test.drop(['id', 'TARGET'],axis = 1)
test.info()
```

```python
test.isnull().sum().sort_values(ascending=False)
```

```python
test['Fld_3'] [539] = test['Fld_3'][539].replace('_', '')
test['Fld_3'][2202] = test['Fld_3'][2202].replace('_', '')
test['Fld_3'] = pd.to_numeric(test['Fld_3'])
test = test.drop(['Fld_408'], axis = 1)
test.fillna(-999, inplace = True)
```

```python
test.isnull().sum().sort_values(ascending=False)
```

```python
from sklearn.preprocessing import MinMaxScaler

scale = MinMaxScaler()
scaled_train = scale.fit_transform(train)
scaled_test = scale.fit_transform(test)
```

```python
scaled_train = pd.DataFrame(scaled_train, columns = train.columns)
scaled_test = pd.DataFrame(scaled_test, columns = test.columns)
```

```python
X = train.drop(['TARGET'], axis = 1)
y = train['TARGET']
```

```python
cate_features_index = np.where(X.dtypes != float)[0]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test,y_train,y_test = train_test_split(X,y, test_size=0.15)
```

```python
from pycaret.regression import compare_models, setup

setup(train, target = 'TARGET')
```

```python
compare_models()
```

```python
from catboost import CatBoostClassifier
from sklearn import metrics
```

```python
cat = CatBoostClassifier(depth = 8, rsm = 0.98, loss_function = 'Logloss', eval_metric = 'Accuracy', use_best_model = True)
cat.fit(X_train, y_train, eval_set=(X_test, y_test), verbose = 200, cat_features = cate_features_index)
pred = cat.predict(X_test)
metrics.accuracy_score(y_test, pred)
```

```python
metrics.precision_score(y_test, pred)
```

```python
metrics.recall_score(y_test, pred)
```

```python
metrics.f1_score(y_test, pred)
```

```python
metrics.confusion_matrix(y_test, pred)
```

```python
cat.score(X_train, y_train)
```

```python
cat.score(X_test, y_test)
```
